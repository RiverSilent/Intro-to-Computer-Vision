```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
%matplotlib inline
```

```python
# Read in the image

img = cv2.imread("img/forest.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
print("The image is ",type(img),"with dimensions ",img.shape)
```

## Task 1: 

#### Replace the background of the STOP sign with a new background image

```python
# Define thresholds:

lower_red = np.array([130,0,0])
upper_red = np.array([255,80,80])
```

```python
img_copy = np.copy(img)

mask = cv2.inRange(img_copy,lower_red,upper_red)
```

```python
plt.imshow(mask,cmap='gray')
```

```python
img_copy[mask == 0] = [0,0,0]
plt.imshow(img_copy)
```

```python
background_img = cv2.imread("img/space.jpg")
background_img = cv2.cvtColor(background_img,cv2.COLOR_BGR2RGB)

cropped_background = background_img[0:450,0:600]

masked_img = np.copy(cropped_background)
```

```python
plt.imshow(masked_img)
```

```python
masked_img[mask != 0] = [0,0,0]
```

```python
plt.imshow(masked_img)
```

#### Hard to see, but the STOP sign is now placed with a new background
