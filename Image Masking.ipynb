## Task: Mask the RED portion of the STOP sign

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

```python
# Read in the image

img = cv2.imread("img/forest.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
# Print out the image statistics

print("The image is ",type(img)," with dimensions ",img.shape)
```

```python
# Define the color thresholds for masking

lower_red = np.array([130,0,0])
upper_red = np.array([255,70,70])
```

```python
img_copy = np.copy(img)
mask = cv2.inRange(img_copy,lower_red,upper_red)
```

```python
plt.imshow(mask,cmap='gray')
```
